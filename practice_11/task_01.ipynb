```python
import numpy as np
from sympy import *
from sympy.plotting import plot
```

```python
x = Symbol('x')
f = -12 * x ** 4 * sin(cos(x)) - 18 * x ** 3 + 5 * x ** 2 + 10 * x - 30
```

```python
def func(x):
    return -12 * x ** 4 * np.sin(np.cos(x)) - 18 * x ** 3 + 5 * x ** 2 + 10 * x - 30
```

```python
def bisection(f, lhs, rhs, eps=1e-6):

  mid = (lhs + rhs) / 2

  while abs(lhs - rhs) > eps:
    mid = (lhs + rhs) / 2

    if f(lhs) * f(mid) > 0: # f(lhs) > 0 and f(mid) > 0 or f(lhs) < 0 and f(mid) < 0
      lhs = mid
    else:
      rhs = mid
      
  return mid
```

```python
print(bisection(func, -12, -8))
print(bisection(func, -8, -6))
print(bisection(func, -6, -2))
print(bisection(func, 0, 3))
print(bisection(func, 3, 5))
print(bisection(func, 5, 10))
print(bisection(func, 10, 12))
```
